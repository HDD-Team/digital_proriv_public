```python
#EEE import
from paddleocr import PaddleOCR, draw_ocr
from matplotlib import pyplot as plt
import cv2
import os
```

```python
#init model
ocr_model = PaddleOCR(lang='en')
```

```python
#init photo
img_path = os.path.join('.', 'drug1.jpg')
```

```python
#get result
result = ocr_model.ocr(img_path)
```

```python
for res in result:
    print(res[1][0]) 
```

ВИЗУАЛ!

```python
boxes = [res[0] for res in result] # 
texts = [res[1][0] for res in result]
scores = [res[1][1] for res in result]
```

```python
#binary img
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

MATPLOTLIB

```python
plt.figure(figsize=(15,15))
annotated = draw_ocr(img, boxes, texts, scores) 
plt.imshow(annotated) 
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load an image
img_path = 'img_paddel.png'
img = Image.open(img_path)

# Perform OCR on the image
results = det_model.ocr(img_path)
print('RES -->',result[0][0][0])

# Check if any text was detected
if len(results) > 0 and len(results[0]) > 0:
    # Plot the detection results
    plt.imshow(img)
    for box in results[0]:
        x1, y1, x2, y2 = [h for h in box[0]]
        plt.gca().add_patch(plt.Rectangle(x1, y1, 30, fill=False, linewidth=2))
    plt.show()
else:
    print("No text was detected in the image.")
```
